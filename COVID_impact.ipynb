```python
import folium
import pandas as pd
```

```python
### import datasets
city_info = pd.read_excel(r'/Users/royliao/Desktop/no.xlsx', sheet_name = 0)
city_number = pd.read_excel(r'/Users/royliao/Desktop/no.xlsx', sheet_name = 1)
city_all = city_info.merge(city_number, left_on='city', right_on='City')
company_info = pd.read_excel(r'/Users/royliao/Desktop/Excel_HÃ¸gskulen_juni21_-21100544.xlsx', sheet_name=3)
company_with_city = company_info.merge(city_all, left_on='KOMNR2', right_on='Number')
```

```python
center_lat = company_with_city.mean().lat
center_long = company_with_city.mean().lng

m = folium.Map(location=[center_lat, center_long], zoom_start=4.5)

for state in company_with_city["city"].unique():
    state_census = company_with_city[company_with_city["city"]==state]
    folium.CircleMarker(
        location=[state_census.lat.values[0], state_census.lng.values[0]],
        radius = float(state_census["Q4"].values[0]),
        popup="COVID affect on sales : %s"%state_census["Q2"].values[0],
        tooltip = state_census.city.values[0],
        color="black",
        fill_color="blue"
    ).add_to(m)

m
```
